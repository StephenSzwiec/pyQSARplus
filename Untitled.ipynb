```python
df_X = []
df_Xex = []
df_Y = []
df_Yex = []
for i in list_of_pro:
    # combined data frame for X and Y
    combined = pd.concat([dfx, i], axis=1)
    # sort by last column
    combined.sort_values(by=combined.columns[-1], inplace=True)
    # split sequentially, every 5 items, to get indices
    test = combined[::5].index
    train = combined.index.difference(test)
    print(test)
    # use indices to generate splits for X and Y
    df_X.append(dfx.loc[train])
    df_Xex.append(dfx.loc[test])
    df_Y.append(i.loc[train])
    df_Yex.append(i.loc[test])
    combined = pd.DataFrame()
```
